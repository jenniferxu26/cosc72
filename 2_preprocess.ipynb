```python
from pathlib import Path
import json, random, re, unicodedata, pandas as pd
from tqdm.auto import tqdm
```

# Setup paths & constants

```python
ROOT = Path(".")
RAW_DIR = ROOT / "data" / "raw"
PROC_DIR = ROOT / "data" / "proc"
PROC_DIR.mkdir(parents=True, exist_ok=True)

RAW_TANG = RAW_DIR / "Tang.json"
RAW_SONG = RAW_DIR / "Song.json"
RAW_GUTEN = RAW_DIR / "gutenberg_en.csv"

SEED = 42
N_PAIRS = 2000
N_GUTEN = 5000
TRAIN_FRAC = 0.80
random.seed(SEED)
```

# Sample bilingual pairs & mono-EN lines

```python
from collections import Counter
from itertools import zip_longest

def normalise(s: str) -> str:
    return unicodedata.normalize("NFKC", s).strip()

def load_poem_lines(fp: Path):
    # return list of dicts {'zh', 'en'} per line
    data = json.load(open(fp, encoding="utf-8"))
    if isinstance(data, dict):
        data = [data]

    rows = []
    for poem in data:
        zh_lines = poem["Chinese"]["content"]
        en_lines = poem["English"]["content"]

        # won't lose a line if one side is longer
        for zh, en in zip_longest(zh_lines, en_lines, fillvalue=""):
            rows.append({"zh": normalise(zh), "en": normalise(en)})
    return rows

rows = (load_poem_lines(RAW_TANG) + load_poem_lines(RAW_SONG))

full_df = pd.DataFrame(rows)
print(f"Loaded {len(full_df):,} (Chinese line, English line) pairs")
```

```python
full_df = full_df.dropna()
full_df
```

```python
avail_pairs = len(full_df)
sample_n    = min(N_PAIRS, avail_pairs)

sample_df = (full_df
             .sample(n=sample_n, random_state=SEED, replace=False)
             .reset_index(drop=True))

guten_df = (pd.read_csv(RAW_GUTEN, encoding="utf-8")      # full file
              .sample(n=N_GUTEN, random_state=SEED, replace=False)
              .reset_index(drop=True))

print(f"Sampling {len(sample_df):,} parallel pairs "
      f"+ {len(guten_df):,} mono-EN lines")
```

# Cleaning functions

```python
# preprocess
PUNCT_MAP = str.maketrans({
    "，": ",", "。": ".", "？": "?", "！": "!", "：": ":", "；": ";",
    "「": "\"", "」": "\"", "『": "\"", "』": "\"",
    "（": "(", "）": ")", "《": "\"", "》": "\"", "—": "-", "﹏": "_"
})
SPLIT_RE = re.compile(r"[，,。.\n]+")

def norm_en(t): return unicodedata.normalize("NFKC", t).strip()
def norm_zh(t): return unicodedata.normalize("NFKC", t.translate(PUNCT_MAP)).strip()
def split_zh(t): return [ln for ln in SPLIT_RE.split(t) if ln]
```

# Build record list

```python
# apply and combine
records = []

for zh_raw, en_raw in tqdm(sample_df[["zh", "en"]].itertuples(index=False),
                           total=len(sample_df), desc="Cleaning bilingual"):
    zh_clean = "\n".join(split_zh(norm_zh(zh_raw)))
    en_clean = norm_en(en_raw)
    records.append({"zh": zh_clean, "en": en_clean, "gloss": ""})

for en_raw in guten_df["text"]:
    records.append({"zh": "", "en": norm_en(en_raw), "gloss": ""})

print(f"Total records ready: {len(records):,}")
```

# Shuffle + split

```python
# split test/train
random.shuffle(records)
cut = int(len(records) * TRAIN_FRAC)
train_recs, test_recs = records[:cut], records[cut:]

def write_jsonl(path: Path, data):
    with path.open("w", encoding="utf-8") as f:
        for row in data:
            json.dump(row, f, ensure_ascii=False)
            f.write("\n")

write_jsonl(PROC_DIR / "train.jsonl", train_recs)
write_jsonl(PROC_DIR / "test.jsonl",  test_recs)

print(f"train.jsonl → {len(train_recs):,} rows")
print(f"test.jsonl  → {len(test_recs):,} rows")
```
